```python
import requests
import jsonify
import pandas as pd
import numpy as np

market = 'BTCUSDT'
tick_interval = '1h'
startTime = 1648771200000
startTime_2 = 1612137600000         # 1 Februari 2021
startTime_3 = 1577836800000         # 1 Januari 2020
endTime = 1649363212000             # 1 April 2022
appended_data = []
df = pd.DataFrame()
while startTime_3<endTime:
    url = 'https://api.binance.me/api/v3/klines?symbol='+market+'&interval='+tick_interval+'&limit=1000'+'&startTime='+str(startTime)+'&endTime='+str(endTime)
    data = requests.get(url).json()
    df = pd.concat([df, pd.DataFrame(data)])
    startTime_3 = data[-1][0] + 3600000
```

```python
df[0].is_unique
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.to_csv('historical_prices/'+market+'_'+tick_interval+'_2022_2022.csv')
```

```python
import requests
import jsonify
import pandas as pd
import numpy as np

market = 'BTCUSDT'
tick_interval = '1d'
startTime = 1502928000000           # 17 Agustus 2017
endTime = 1648789200000             # 1 April 2022
appended_data = []
df = pd.DataFrame()
while startTime<endTime:
    url = 'https://api.binance.me/api/v3/klines?symbol='+market+'&interval='+tick_interval+'&limit=1000'+'&startTime='+str(startTime)+'&endTime='+str(endTime)
    data = requests.get(url).json()
    df = pd.concat([df, pd.DataFrame(data)])
    startTime = data[-1][0] + 86400000
```
